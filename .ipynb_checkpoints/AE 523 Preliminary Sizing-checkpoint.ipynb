```python
#Initializing for graphs and needed packages
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

#Graph Formatting
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 14
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size
Res=100 #Resoultion of the calulations

#Graph Font Edit
font = {'size'   : 22}
matplotlib.rc('font', **font)
```

## Solar Cell Generation Estimate

```python
#Area as a Function of Distance from the Sun
#Distance from Sun (km*10^6)
RR=np.linspace(50,250,Res)

#Static Parameters
min=0.149              #Minimum efficiency of 15% at 28 C
Psa=4000                #Power the cells need to provide | W
cell=8                  #Cell Size | 2 x 4 Silicon Cell
SF=1370                 #Solar Flux at Earth | W/m^2
PF=0.88                 #Packing Factor
PL=(SF/10000)*min*cell #Power output of Minimum Cell | W/cell

#Calculating cell | Lumped Cell Efficiency
rad=0.7        #Power Loss from Radiation Damage after 10 Years
uv=0.98        #Power Loss from UV Discoloration
cy=0.99        #Power Loss by Thermal Cycling
m=0.975        #Power Loss by Cell Mismatch
l=0.98         #Power Loss by Resistance in Cell Interconnects
con=0.99       #Power Loss Caused by Contamination from All Sources
=np.deg2rad(5) #Pointing System Controls Within 5 Deg
s=1            #Power Loss Caused by Shadowing (Designed)
Lp=np.cos()    #Array Pointing Loss Factor
R=149.6         #Distance from Sun in Mil. Kilometers
HI=(149.6/RR)**2#Adjustment of Solar Intensity
T=115           #Temperature of Cell in Orbit | C |-120-115C
t=28            #Temperature the Cells Were Tested | C
t=1-0.005*(T-t)#Power Adjustment for Operation Temperature
cell=(rad*uv*s*cy*HI*m*l*con*t*Lp) #Lumped Cell Efficiency
Pc=PL*(cell)   #Power of One Cell Under Operational Conditions

#Assembly
Nc=Psa/Pc       #Number of cells needed for Power | cells
Dc=(10000/8)*PF #Cell Density | cells/m^2
Asa=Nc/Dc       #Total Area of the Array | m^2

#Outputs
plt.subplot(2,1,1)
plt.plot(RR,Asa)
plt.axvline(x=57.8, label="Mercury",color="gray")
plt.axvline(x=108.2, label="Venus",color="pink")
plt.axvline(x=149.6, label="Earth",color="green")
plt.axvline(x=228.0, label="Mars",color="red")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.xlabel("Distance from the Sun (millions of km)")
plt.ylabel("Solar Array Area $(m^2)$")
plt.title("Sun Proximity Effect on Solar Array Area to Mars")

plt.tight_layout()

plt.show()
```

#### Solar Array Conclusion
Solar panels are roughly half as effective at Mars compared to Earth. Power generation therefore needs to be sized for the the minimum or for the case. Under sizing and then relying on battery power is not an option because of the transfer time and potential for the spacecraft to stay with the Mars gateway for an extended time.
* This is a safe option and very effective at the distances from Earth
* The mission will over produce energy for the begining of the mission so it is possible for two solar arrays and only one deployed until ~185 km from the Sun.

#### Space Shuttle Electrical Power Systems
https://spaceflight.nasa.gov/shuttle/reference/shutref/orbiter/eps/overview.html
- The space shuttle uses hydrogen for power generation with limited byproducts.
- 28 V DC power is generated then converted to AC power
> The three main dc buses-MNA, MNB and MNC-are the prime sources of power for the vehicle's dc loads. Each of the three dc main buses supplies power to three solid-state (static), single-phase inverters, which constitute one three-phase alternating-current bus; thus, the nine inverters convert dc power to 115-volt, 400-hertz ac power for distribution to three ac buses-AC1, AC2 and AC3-for the vehicle's ac loads.
- The "Assemblies" are used to guide and control power.
>The EPDC subsystem controls and distributes electrical power (ac and dc) to the orbiter subsystems, the solid rocket boosters, the external tank and payloads. Power is controlled and distributed by assemblies. Each assembly is a housing for electrical components, such as remote switching devices, buses, resistors, diodes and fuses. Each assembly usually contains a power bus or buses and remote switching devices for distributing bus power to subsystems located in its area.
